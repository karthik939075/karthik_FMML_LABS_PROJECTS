### Example 1 - A simple cost function

#### $$f(x) = x^2 + x + 1$$

````python
def f(x):
    return x ** 2 + x + 1
````

````python
# Make Data
x_1 = np.linspace(start=-3, stop=3, num=500)
````

````python
# Plot
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
````

### Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

````python
def df(x):
    return 2 * x + 1
````

````python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
````

### Gradient Descent

````python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # Appending the slope (df(new_x)) to slope_list

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

````python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

### Example 3 - Divergence and Overflow

#### $$h(x) = x^5 - 2x^4 + 2$$

````python
import numpy as np
import matplotlib.pyplot as plt

# Make some data
x_2 = np.linspace(-2, 2, 1000)

# Define the function g(x)
def g(x):
    return x**2 + 2*x + 1  # Simple quadratic function

# Define the derivative of the function g'(x)
def dg(x):
    return 2*x + 2  # Derivative of g(x) = 2x + 2

# Gradient Descent Setup
new_x = 3  # Starting point
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [dg(new_x)]  # We are using dg(x) for gradient descent

for n in range(500):
    previous_x = new_x
    gradient = dg(previous_x)  # Calculate the gradient (derivative)
    new_x = previous_x - step_multiplier * gradient  # Update the x value

    step_size = abs(new_x - previous_x)  # Measure step size
    x_list.append(new_x)
    slope_list.append(dg(new_x))  # Append the slope (derivative)

    # Check for convergence
    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or dg(x) value at this point is:', dg(new_x))
print('f(x) value or cost at this point is:', g(new_x))

# Plotting the function and its derivative side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the function g(x)
ax1.plot(x_2, g(x_2), label='g(x) = x^2 + 2x + 1', color='b')
ax1.set_title('Function g(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('g(x)')
ax1.grid(True)
ax1.legend()

# Plot the derivative g'(x)
ax2.plot(x_2, dg(x_2), label="g'(x) = 2x + 2", color='r')
ax2.set_title('Derivative g\'(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('g\'(x)')
ax2.grid(True)
ax2.legend()

# Show the plot
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt

# Assuming dg is the derivative of the function you are optimizing
def dg(x):
    # Example: Derivative of f(x) = x^2 (just for illustration)
    return 2 * x

# Assuming gradient_descent is already defined and works as expected
def gradient_descent(derivative_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    history = [x]
    for i in range(max_iter):
        grad = derivative_func(x)
        x = x - multiplier * grad  # Gradient descent step
        history.append(x)
        if abs(grad) < precision:  # Stop if the gradient is small enough
            break
    return history

# Perform gradient descent with low_gamma, mid_gamma, high_gamma, and insane_gamma
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)
mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n)
high_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n)
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=n)

# Plot the results
plt.figure(figsize=(10, 6))

# Plot all the results
plt.plot(low_gamma, label='Low gamma (0.0005)', color='blue')
plt.plot(mid_gamma, label='Mid gamma (0.001)', color='green')
plt.plot(high_gamma, label='High gamma (0.002)', color='red')
plt.plot(insane_gamma, label='Insane gamma (0.25)', color='purple')

# Add labels and title
plt.xlabel('Iterations')
plt.ylabel('Value of x')
plt.title('Gradient Descent with Different Learning Rates')
plt.legend()

# Show the plot
plt.show()
````

````python
import matplotlib.pyplot as plt

# Define the cost function (J(x) = x^2 for the example)
def cost_function(x):
    return x ** 2

# Modified gradient descent to track cost reduction
def gradient_descent_with_cost(derivative_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    history_x = [x]  # To store the values of x at each iteration
    cost_history = [cost_function(x)]  # To store the cost (J(x)) at each iteration

    for i in range(max_iter):
        grad = derivative_func(x)
        x = x - multiplier * grad  # Gradient descent step
        history_x.append(x)
        cost_history.append(cost_function(x))  # Track cost at each iteration

        if abs(grad) < precision:  # Stop if the gradient is small enough
            break

    return history_x, cost_history

# Perform gradient descent with different learning rates
n = 100
low_gamma_x, low_gamma_cost = gradient_descent_with_cost(derivative_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)
mid_gamma_x, mid_gamma_cost = gradient_descent_with_cost(derivative_func=dg, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n)
high_gamma_x, high_gamma_cost = gradient_descent_with_cost(derivative_func=dg, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n)
insane_gamma_x, insane_gamma_cost = gradient_descent_with_cost(derivative_func=dg, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=n)

# Plot the reduction in cost for each iteration
plt.figure(figsize=(10, 6))

# Plot cost reduction over iterations for each gamma
plt.plot(low_gamma_cost, label='Low gamma (0.0005)', color='blue')
plt.plot(mid_gamma_cost, label='Mid gamma (0.001)', color='green')
plt.plot(high_gamma_cost, label='High gamma (0.002)', color='red')
plt.plot(insane_gamma_cost, label='Insane gamma (0.25)', color='purple')

# Add labels and title
plt.xlabel('Iterations')
plt.ylabel('Cost (J(x))')
plt.title('Cost Reduction with Different Learning Rates')
plt.legend()

# Show the plot
plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
